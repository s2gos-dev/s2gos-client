```python
import panel as pn

pn.extension()

# Define some widgets
algo = pn.widgets.Select(name='Algorithm', options=['NDVI', 'EVI', 'LAI'])
threshold = pn.widgets.FloatSlider(name='Threshold', start=0.0, end=1.0, step=0.01, value=0.3)
submit = pn.widgets.Button(name='Submit Job', button_type='primary')
output = pn.pane.Markdown("")

# Define the callback
def on_submit(event):
    output.object = (
        f"🚀 Submitted job with:\n"
        f"- Algorithm: **{algo.value}**\n"
        f"- Threshold: **{threshold.value}**"
    )

submit.on_click(on_submit)

# Layout
form = pn.Column(
    "## S2GOS Scene Generator",
    algo,
    threshold,
    submit,
    output,
)

form.servable()
```
